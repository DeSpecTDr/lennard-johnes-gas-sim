```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 20)
y = x ** 2 + np.random.random(20)


plt.plot(x, y, '.')

n = 10
xx = x[-n:]
yy = y[-n:]
k, b = np.polyfit(xx, yy, 1)
mi = np.min(xx)
ma = np.max(xx)
mima = 0.05 * (ma - mi)
xxx = np.array([mi - mima, ma + mima])
yyy = k * xxx + b

plt.plot(xxx, yyy)
```
